```python
import pandas as pd
import numpy as np
import geopandas
import os
%matplotlib inline
```

```python
poorHousing = pd.read_csv("housing-conditions-beller/poorhousingconditions.csv")
neighborhoods = geopandas.read_file("Neighborhoods/Neighborhoods_.shp")
```

```python
neighborhoods['tractce10'] = neighborhoods.tractce10.replace(' ',np.nan).astype(float)
neighborhoods['tractce10'] = pd.to_numeric(neighborhoods['tractce10'])
```

```python
poorHousingMap = neighborhoods.merge(poorHousing, how='left', left_on='tractce10', right_on='TRACT')
```

```python
poorHousingMap = GeoDataFrame(poorHousingMap)
```

```python
poorHousingMap.plot(column='PPoorCon', # set the data to be used for coloring
               cmap='RdYlGn_r',              # choose a color palette
               edgecolor="white",        # outline the districts in white
               legend=True,              # show the legend
               legend_kwds={'label': "Proportion of Houses in Poor Condition"}, # label the legend
               figsize=(15, 10),         # set the size
               missing_kwds={"color": "lightgrey"} # set disctricts with no data to gray
               )
```
